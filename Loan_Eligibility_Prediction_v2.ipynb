# IMPORTING DATASETS

```python
df = pd.read_csv("Loan_Eligibility.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
df.iloc[:,0] = LabelEncoder().fit_transform(df.iloc[:,0].astype('str'))
df.iloc[:,1] = LabelEncoder().fit_transform(df.iloc[:,1].astype('str'))
df.iloc[:,2] = LabelEncoder().fit_transform(df.iloc[:,2].astype('str'))
df.iloc[:,3] = LabelEncoder().fit_transform(df.iloc[:,3].astype('str'))
df.iloc[:,4] = LabelEncoder().fit_transform(df.iloc[:,4].astype('str'))
df.iloc[:,5] = LabelEncoder().fit_transform(df.iloc[:,5].astype('str'))
df.iloc[:,11] = LabelEncoder().fit_transform(df.iloc[:,11].astype('str'))
df.iloc[:,12] = LabelEncoder().fit_transform(df.iloc[:,12].astype('str'))

df
```

```python
dataset=df.values
dataset
```

```python
X = dataset[:,1:12]
Y = dataset[:,12]
```

# PreProcessing Data

```python
from sklearn import preprocessing
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
X_scale
```

```python
import matplotlib.pyplot as plt
sns.heatmap(df.corr(),annot = True,cmap='magma')
fig=plt.gcf()
fig.set_size_inches(20,20)
plt.show()
```

# Spliting into Train & Test Data

```python
from sklearn.model_selection import train_test_split
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.3)
```

```python
x_val, X_test, y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)
```

```python
print(X_train.shape, X_test.shape, x_val.shape, Y_train.shape, Y_test.shape, y_val.shape)
```

# Neural Network Desigh

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras import regularizers
```

```python
model = Sequential([
    Dense(32, activation='relu', input_shape=(11,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid'),
])
```

```python
model.compile(optimizer='sgd',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
hist=model.fit(X_train, Y_train, 
          batch_size = 32, epochs = 100,
        validation_data = (x_val, y_val))
```

```python
model.evaluate(X_test, Y_test)[1]
```

# Visualizing Loss and Accuracy

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('LOSS')
plt.xlabel('EPOCH')
plt.legend(['train', 'val'], loc='upper right')
plt.show()
```

```python
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='lower right')
plt.show()
```
